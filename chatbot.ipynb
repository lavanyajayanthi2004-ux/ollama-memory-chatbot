```python
# imports
import os
import json
from dotenv import load_dotenv
from IPython.display import Markdown, display, update_display
from scraper import fetch_website_links, fetch_website_contents
from openai import OpenAI
```

```python
# constants

# MODEL_GPT = 'gpt-4o-mini'
MODEL_LLAMA = 'llama3.2'
```

```python
# set up environment
OLLAMA_BASE_URL='http://localhost:11434/v1'
openai=OpenAI(base_url=OLLAMA_BASE_URL,api_key='ollama')
```

```python
run=True
input_question=input()
question = f"{input_question}"
system_prompt="""You are a friendly, humorous assistant who answers user questions clearly while keeping the tone light, engaging, and enjoyable."""
user_prompt=f"""Here is {question}.Answer for the question clearly and accurately, using light humor to keep things fun and easy to understand."""
system_messages=[{"role":"system","content":system_prompt},{"role":"user","content":user_prompt}]
while run:
    
    response=openai.chat.completions.create(model=MODEL_LLAMA,messages=system_messages)
    result=response.choices[0].message.content
    print("assistant: ",result)
    again=input("Do you have any other questions..if so type 'yes' else type 'no' ").lower()    
    system_messages.append({"role":"assistant","content":result})
    again_question=input()
    print("user: ",again_question)
    system_messages.append({"role":"user","content":again_question})
    if again!="yes":
        run=False

    
```
